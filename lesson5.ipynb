# Data Visualization via Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Line Plots

```python
# Plotting the sqaure root of 20 equidistance numbers between 0 and 20 
import numpy as np
import math
x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.plot(x_vals,y_vals)
```

```python
# A different way to plot the same graph

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]

fig = plt.figure()
ax = plt.axes()
ax.plot(x_vals,y_vals)
```

```python
#Changing the plot size

plt.rcParams["figure.figsize"] = [8,6]

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.plot(x_vals,y_vals)
```

# Titles, labels, and Legends

```python
# Adding labels and a title to the graph

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Square Roots")
plt.plot(x_vals,y_vals)
```

```python
# Changing the color of the plot

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Square Roots")
plt.plot(x_vals,y_vals,"r")
```

```python
# Adding a legend to the plot

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Square Roots")
plt.plot(x_vals,y_vals,"r",label="Square Root")
plt.legend(loc="upper center")
```

```python
#PLotting multiple line plots inside one graph


x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
y2_vals = x_vals ** 3

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Square Roots")
plt.plot(x_vals,y_vals,"r",label="Square Root")
plt.plot(x_vals,y2_vals,"b",label="Cube")
plt.legend(loc="upper center")
```

# Data Visualization via Seaborn

```python
import seaborn as sns

plt.rcParams["figure.figsize"] = [10,8]
tips_data = sns.load_dataset("tips")
tips_data.head()
```

# The Dist Plot

```python
# Using the dist plot for the total_bill column of the tips dataset

plt.rcParams["figure.figsize"] = [10,8]
sns.distplot(tips_data["total_bill"])
```

# The joint Plot

```python
#Using a joint plot for total bill and tip

sns.jointplot(x="total_bill",y="tip",data=tips_data)
```

```python
# Replacing the scatter plot with a regression line 
sns.jointplot(y="total_bill",x="size",data=tips_data,kind="reg")
```

# The pair plot

```python
sns.pairplot(data=tips_data)
```

# The Bar Plot

```python
plt.rcParams["figure.figsize"] = [8,6]
sns.set_style("darkgrid")

titanic_data = sns.load_dataset("titanic")
titanic_data.head()
```

```python
#Using a bar plot to show the average age of passenagers traveling in the different classes on the ship

sns.barplot(x="pclass",y="age",data=titanic_data)
```

# The Count Plot

```python
#Count plot for the pclass column of the titanic dataset

sns.countplot(x="pclass",data=titanic_data)
```

# The Box Plot

```python
#Using a box plot for the fare column of the titanic dataset

sns.boxplot(x=titanic_data["fare"])
```

# The Violin Plot

```python
#Violin plot for the passenagers traveling alone and for the passengers traveling along with another passenager

sns.violinplot(x="alone",y="age",data=titanic_data)
```
